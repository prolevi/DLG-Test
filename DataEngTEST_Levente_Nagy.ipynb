```python
## Because both .csv files look similar in structure, same number of columns and same table header, is very easy to concatenate the dataframes I created so far (axis=0 - it means rows).
vertical_stack = pd.concat([weatherdlg1_df, weatherdlg2_df], axis=0)
```

```python
## Now that all .csv files are in a single dataframe I can export it to a .parquet file using pyarrow.
import pyarrow as pa
```

```python
## I can convert the dataframe in a parquet file using 'snappy' compression. The parquet file name contains the compression type.
vertical_stack.to_parquet('weatherall.snappy.parquet' ,engine='auto', compression='snappy')
```

```python
## I can also convert the dataframe using a different compression type: brotli.
vertical_stack.to_parquet('weatherall.brotli.parquet' ,engine='auto', compression='brotli')
```

```python
## We can see the parquet file compressed using "brotli" is smaller in size than "snappy" compression type file. You can also compare with the 2 .csv file from the beginning of the test, please see image below.
from IPython.display import Image
Image(filename="filesize.JPG", width=700, height=700)
```
